# Desenvolver um programa que leia e exiba o nome de cinco estados brasileiros e suas respectivas capitais.
*    Exemplo:
*    Estado - Capital
*    SP     - São Paulo
*    BA     - Salvador
*    AL     - Maceio
*    AC     - Rio Branco
*    TO     - Palmas

```python
# 06/10/2023
# Brennon Meireles
# Senai Suiço Brasileiro
# Exercício de FPOO

brasil = [] # lista principal

#   loop para armazenar o conteudo para a Lista principal:
for i in range(5):
  lista  = []  # lista auxiliar

  estado = str(input('Digite o Estado: ')) # input para receber os dados
  capital = str(input('Digite a Capital do estado: '))  # input para receber os dados

  lista.append(estado) # adição da var estado para lista
  lista.append(capital) # adição da var capital para lista

  brasil.append(lista) # adição de todo o conteudo da lista aux para a lista principal

# Saida :
for listas in brasil:
  estado, capital = listas
  print(f'Estado: {estado}, Capital: {capital}')
```

```python
# versão otimizada:

brasil = []

for _ in range(5):
    estado = input('Digite o Estado: ')
    capital = input('Digite a Capital do estado: ')
    brasil.append([estado, capital])

# Saida:
for estado, capital in brasil:
    print(f'Estado: {estado}, Capital: {capital}')
```
